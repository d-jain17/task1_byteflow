```python
import pandas as pd
```

```python
df=pd.read_csv('data1.csv')
print(df)
```

```python
#reading headers
print(df.columns)
```

```python
#reading name column
print(df['Name'])
```

```python
#getting first 5 entries of name column
print(df['Name'][0:5])
```

```python
#reading name , sex and age
print(df[['Name','Sex','Age']])
```

```python
#reading first four rows
print(df.head(4))
```

```python
#sorting data according to order of surnames
df.sort_values('Name')
```

```python
#sorting data according to sex
df.sort_values('Sex')
```

```python
#sorting data according to increasing order of age
df.sort_values('Age')
```

```python
#sorting data according to decreasing order of age
df.sort_values('Age',ascending=False)
```

```python
#sorting data according to order of passengerId
df.sort_values('PassengerId')
```

```python
#sorting all females in increasing order of age and then males in increasing order of age
df.sort_values(['Sex','Age'])
```

```python
#sorting all females in increasing order of passengerId and then males in increasing order of passengerId
df.sort_values(['Sex','PassengerId'])
```

```python
#sorting all males in increasing order of passengerId and then females in increasing order of passengerId
df.sort_values(['Sex','PassengerId'],ascending=[0,1])
```

```python
#sorting all males in decreasing order of passengerId and then females in decreasing order of passengerId
df.sort_values(['Sex','PassengerId'],ascending=[0,0])
```

```python
#filtering all the male passengers
df.loc[df['Sex']=='male']
```

```python
#filtering all the female passengers
df.loc[df['Sex']=='female']
```

```python
#filtering all the male passengers with age below 20
df.loc[(df['Sex']=='male') & (df['Age']<20)]
```

```python
#filtering all the female passengers with age above 20
df.loc[(df['Sex']=='female') & (df['Age']>20)]
```

```python
#filtering all the male passengers with passengerId below 20
df.loc[(df['Sex']=='male') & (df['PassengerId']<20)]
```

```python
#filtering all the male passengers with age below 20 who survived
df.loc[(df['Sex']=='male') & (df['Age']<20) & (df['Survived']>0)]
```

```python
#filtering all the female passengers with age above 20 who survived
df.loc[(df['Sex']=='female') & (df['Age']>20) & (df['Survived']>0)]
```

```python
#filtering all the male passengers with age below 20 whose pclass is 1
df.loc[(df['Sex']=='male') & (df['Age']<20) & (df['Pclass']==1)]
```

```python
#filtering all the male passengers with Pclass1 who survived
df.loc[(df['Sex']=='male') & (df['Pclass']==1) & (df['Survived']>0)]
```

```python
# filtering all the male passengers with Pclass 1 or who survived
df.loc[(df['Sex']=='male') & ((df['Pclass']==1) | (df['Survived']>0))]
```

```python
# filtering all the passengers with Pclass 1 who survived or are female
df.loc[(df['Pclass']==1) & ((df['Sex']=='female') | (df['Survived']>0))]
```
